```python
# First, install required AWS packages
!pip install boto3
!pip install awscli

import os
import boto3
from google.colab import userdata
import json

def configure_aws_colab(aws_access_key_id, aws_secret_access_key, region='us-east-1'):
    """
    Configure AWS credentials in Google Colab environment

    Parameters:
    aws_access_key_id (str): AWS access key ID
    aws_secret_access_key (str): AWS secret access key
    region (str): AWS region name (default: 'us-east-1')
    """

    # Set AWS credentials as environment variables
    os.environ['AWS_ACCESS_KEY_ID'] = aws_access_key_id
    os.environ['AWS_SECRET_ACCESS_KEY'] = aws_secret_access_key
    os.environ['AWS_DEFAULT_REGION'] = region

    # Verify the configuration
    try:
        # Create a boto3 session
        session = boto3.Session(
            aws_access_key_id=aws_access_key_id,
            aws_secret_access_key=aws_secret_access_key,
            region_name=region
        )

        # Test the configuration
        s3 = session.client('s3')
        buckets = s3.list_buckets()
        print("AWS configuration successful!")
        print(f"Number of S3 buckets: {len(buckets['Buckets'])}")
        return True

    except Exception as e:
        print(f"Error configuring AWS: {str(e)}")
        return False

# Example usage with user input (more secure than hardcoding)
from getpass import getpass

# Get AWS credentials securely
aws_access_key = getpass('Enter your AWS Access Key ID: ')
aws_secret_key = getpass('Enter your AWS Secret Access Key: ')
region = input('Enter AWS region (press Enter for us-east-1): ') or 'us-east-1'

# Configure AWS
configure_aws_colab(aws_access_key, aws_secret_key, region)
```
